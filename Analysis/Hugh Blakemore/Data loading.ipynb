```python
#Loading in data
import pandas as pd
url='https://raw.githubusercontent.com/data301-2020-winter2/course-project-group_1017/main/Data/Raw/medical_expenses.csv'
data=pd.read_csv(url)
data
```

```python
data.dropna(subset=['charges'], inplace=True)
data
#drops any data points where charges are not found as main comparisons will be drawn from there
```

```python
data.describe(include='all')# this looks like a very clean dataset 
```

```python
data.rename({'children':'Dependents'},axis=1,inplace=True)# renaming to dependents as makes more sense for this data set as it describes how many people are dependent on this persons medical insurance
data
```

```python
#im interested in analysis of all ths data and how the variesty of the data such as bmi affects medical charges thus i wont be removing any data from this dataset
data['region'].replace('southwest','SW',inplace=True)
data['region'].replace('northwest','NW',inplace=True)
data['region'].replace('southeast','SE',inplace=True)
data['region'].replace('northeast','NE',inplace=True)
data
```

```python
data.charges = data['charges'].round(decimals=2)
data.bmi = data['bmi'].round(decimals=1)
data#changes columns to desired amount if decimal places
```

```python
data.sort_values('charges',inplace=True, ascending=True)
data#sorts data by charges
```

```python
data.reset_index(drop=True,inplace=True)                
data
```

**Step 2 of method chaining**

```python
def load_and_process(url):
    df1 = (
          pd.read_csv(url)
          .rename({'children':'Dependents'},axis=1)
          .dropna(subset=['charges'])
          .replace({'southwest':'SW','southeast':'SE','northeast':'NE','northwest':'NW'})
      )
    df2=(df1
         .round({"charges":2,"bmi":1})
         .sort_values('charges',ascending=True)
         .reset_index(drop=True)  
    )
    return df2
```

```python
load_and_process(url)
```
